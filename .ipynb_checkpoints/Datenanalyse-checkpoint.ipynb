```python
import pandas as pd
```

```python
dataset = pd.read_csv('IT Salary Survey EU  2020.csv')
```

```python
dataset.drop(['Timestamp', 'Years of experience in Germany', 'Yearly bonus + stocks in EUR',
              'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country', 
              'Annual bonus+stocks one year ago. Only answer if staying in same country', 
              'Employment status', 'Сontract duration', 
              'Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week', 
              'Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'], 
             axis=1, inplace=True)
```

```python
print(dataset.columns)
print(dataset)
```

```python
pd.set_option("display.max_rows", None)
print(dataset['Position '].value_counts())
```

```python
mask = dataset['Position '].str.contains('UI/UX', na=False)
dataset.loc[~mask,'Position '] = dataset.loc[~mask,'Position '].str.split(r"\s*/\s*|\s*,\s*")
dataset['Positions'] = dataset['Position '].apply(lambda x: x if isinstance(x, list) else [x])
dataset.explode('Positions')
print(dataset['Positions'].value_counts())
```
